# 🧮 حل مشروع رياضيات وإحصاء علم البيانات (Mathematics & Statistics Solution)

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
```

### 📐 الجزء الأول: الجبر الخطي

```python
# Manual Matrix Multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Using NumPy
print("NumPy multiplication:")
print(np.dot(A, B))

# Determinant and Inverse
det_A = np.linalg.det(A)
inv_A = np.linalg.inv(A)
print(f"Det: {det_A:.2f}, Inverse:\n{inv_A}")
```

### 📈 الجزء الثاني: الانحدار التدريجي (Gradient Descent)

```python
x = 10.0
learning_rate = 0.1
for i in range(50):
    gradient = 2 * x - 4
    x = x - learning_rate * gradient
print(f"Gradient Descent converged to x = {x:.4f} (True minimum is 2.0)")

# Information Theory
# Entropy: H(X) = -sum(p_i * log2(p_i))
p = [0.2, 0.8]
entropy = -sum(pi * np.log2(pi) for pi in p)
print(f"Entropy: {entropy:.4f} bits")
```

### 💬 الجزء الثالث: الإحصاء الاستدلالي واختبار الفرضيات

```python
group_a = [23, 25, 28, 30, 21, 22, 29, 27, 26, 24]
group_b = [31, 33, 29, 35, 34, 30, 32, 28, 30, 32]

t_stat, p_val = stats.ttest_ind(group_a, group_b)
print(f"T-statistic: {t_stat:.4f}, p-value: {p_val:.6f}")
if p_val < 0.05:
    print("Statistically significant difference found (Reject Null Hypothesis)")
else:
    print("No significant difference found (Accept Null Hypothesis)")
```
