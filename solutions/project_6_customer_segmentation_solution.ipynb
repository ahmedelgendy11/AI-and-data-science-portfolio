# 👥 حل مشروع تقسيم فئات العملاء (Customer Segmentation Solution)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

df = pd.read_csv('datasets/customer_segmentation_data.csv')
print(df.head())
```

### 📊 الجزء الأول: طريقة المرفق (Elbow Method)

```python
X = df[['age', 'annual_income_k', 'spending_score_1_100']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertias = []
K_range = range(1, 11)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.tight_layout()
plt.savefig('solutions/customer_segmentation_elbow.png')
plt.show()
```

### 🤖 الجزء الثاني: التدريب وتقليل الأبعاد باستخدام PCA

```python
# We select k=4 from the Elbow plot
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_transform(X_scaled).argmin(axis=1) # or fit_predict
# wait,kmeans.fit_predict(X_scaled) is cleaner
df['Cluster'] = kmeans.fit_predict(X_scaled)

pca = PCA(n_components=2)
pca_res = pca.fit_transform(X_scaled)
df['PCA1'] = pca_res[:, 0]
df['PCA2'] = pca_res[:, 1]

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set1', s=60)
plt.title('Customer Segments Visualized via PCA (2D)')
plt.tight_layout()
plt.savefig('solutions/customer_segmentation_clusters.png')
plt.show()

# Profiles
print("Average values per cluster:")
print(df.groupby('Cluster')[['age', 'annual_income_k', 'spending_score_1_100']].mean())
```
