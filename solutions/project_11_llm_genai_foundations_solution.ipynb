# 🤖 حل مشروع الذكاء الاصطناعي التوليدي ونظام الـ RAG (GenAI & RAG Solution)

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

### 🛠️ الجزء الأول: محاكاة استدعاء نموذج لغوي

```python
def mock_llm_summarize(text):
    prompt = f"""
    Prompt: Summarize the following review into 3 keywords.
    Review: {text}
    Summary:
    """
    # Simulating LLM API response
    lower = text.lower()
    if "laptop" in lower:
        return "Laptop, Fast, Great"
    elif "watch" in lower:
        return "Watch, Broken, Avoid"
    return "Product, Average, Ok"

print(mock_llm_summarize("I absolutely love this laptop, it is super fast and perfect."))
```

### 🤖 الجزء الثاني: بناء نظام RAG للبحث النصي المتجهي

```python
reviews = [
    "I absolutely love this laptop, it is super fast and perfect.",
    "Avoid this watch, it broke down after only two days of use.",
    "The food mixer is okay, nothing special but works fine.",
    "Worst product ever, arrived damaged and extremely cheap quality."
]

query = "looking for a fast computer or laptop"

# TF-IDF Vectorizer to act as simple embeddings
vectorizer = TfidfVectorizer()
transformed_reviews = vectorizer.fit_transform(reviews)
transformed_query = vectorizer.transform([query])

# Compute cosine similarities
similarities = cosine_similarity(transformed_query, transformed_reviews)[0]
best_idx = np.argmax(similarities)

print(f"Similarities: {similarities}")
print(f"Most relevant context retrieved (RAG): \n\"{reviews[best_idx]}\"")

# Simulating the generation stage of RAG
prompt_to_llm = f"""
System: You are a helpful assistant. Use this retrieved context to answer the user query.
Context: {reviews[best_idx]}
User: What is the benefit of the recommended product?
AI Answer:
"""
print("--- Generated Prompt for LLM Context ---")
print(prompt_to_llm)
```
