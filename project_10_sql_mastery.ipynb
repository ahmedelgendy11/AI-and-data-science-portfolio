### 📊 الجزء الأول: عمليات الربط والتجميع الأساسية
اكتب استعلام SQL يدمج المبيعات مع أسماء الموظفين لحساب إجمالي مبيعات كل موظف مرتبة تنازلياً.

```python
# TODO: Query sales and employee names
query = """
SELECT e.name, SUM(s.amount) as total_sales
FROM employees e
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY e.name
ORDER BY total_sales DESC
"""
pd.read_sql_query(query, conn)
```

### 🛠️ الجزء الثاني: الدوال التحليلية المتقدمة (Window Functions)
استخدم دالة `ROW_NUMBER()` أو `RANK()` داخل الاستعلام لترتيب رواتب الموظفين داخل كل قسم إداري بشكل تنازلي.

```python
# TODO: Write SQL query with Window Function ROW_NUMBER() OVER(PARTITION BY...)
query = """
SELECT 
    name,
    dept_id,
    salary,
    ROW_NUMBER() OVER(PARTITION BY dept_id ORDER BY salary DESC) as salary_rank
FROM employees
"""
pd.read_sql_query(query, conn)
```
